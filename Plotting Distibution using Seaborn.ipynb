```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Market=pd.read_csv("D:/MBA/SEM 2/Python/Datasets/market_fact.csv")
```

```python
Market
```

#### Histogram

```python
plt.hist(Market['Sales'])
plt.show
```

#### To view all data in Histogram

```python
plt.hist(Market['Sales'])
plt.yscale('log')
plt.show
```

#### Scatterplot

```python
plt.scatter(Market['Sales'],Market['Profit'])
plt.show()
```

#### To view the density graph on the histogram

```python
sns.distplot(Market['Shipping_Cost'])
plt.show()
```

#### To include rug 

```python
sns.distplot(Market['Shipping_Cost'][:200],rug = True)
plt.show()
```

#### To create density plot alone

```python
sns.distplot(Market['Sales'],hist=False)
plt.show()
```

```python
plt.subplot(2,2,1)
plt.title('Sales')
sns.distplot(Market['Sales'])

plt.subplot(2,2,2)
plt.title('Profit')
sns.distplot(Market['Profit'])



plt.subplot(2,2,3)
plt.title('Order_Quantity')
sns.distplot(Market['Order_Quantity'])

plt.subplot(2,2,4)
plt.title('Shipping_Cost')
sns.distplot(Market['Shipping_Cost'])

plt.tight_layout()

plt.show()
```

#### Bxplot

```python
sns.boxplot(Market['Order_Quantity'])
plt.title('Order Quantity')
plt.show()
```

#### To change axis into horizontal

```python
sns.boxplot(x=Market['Order_Quantity'])
plt.title('Order Quantity')
plt.show()
```

#### Visualizing Bivariate Distribution

```python
sns.jointplot(x='Sales',y='Profit',data=Market)
plt.show()
```

#### Filtering the dataset to have better view

```python
Market1 = Market[(Market.Profit < 10000) & (Market.Sales < 20000)]

sns.jointplot(x='Sales',y='Profit',data=Market1)
plt.show()
```

```python
Market2 = Market[(Market.Profit < 100) & (Market.Profit > -100) & (Market.Sales < 200)]

sns.jointplot(x='Sales',y='Profit',data=Market2,kind="hex",color="g")
plt.show()
```
